# Decision Trees

From the outset you should know that Decision Trees have some pros and cons. 

Pro: They are easily explainable. 
Con: They're not very accurate.

The fact that they're not accurate is why they're rarely used (not surprisingly). 

So why cover decision trees at all? 

Because the Random Forest model is based on Decision Trees, and addresses these issues. It goes without saying that you need to cover Decision Trees before progressing on to Random Forests.

# Decision Tree Example

This example has been taken from.

https://www.w3schools.com/python/python_ml_decision_tree.asp

The data for this tutorial determines if someone would like to go to a comedy show or not. 

```python
import pandas as pd

df = pd.read_csv("../../Data/w3schools_decision_tree.csv")
df
```

```python
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)
df
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features].values
y = df['Go'].values
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

plt.figure(figsize=(12,12))
tree.plot_tree(dtree, feature_names=features)
```

```python
print(dtree.predict([[40, 10, 7, 1]]))

```
